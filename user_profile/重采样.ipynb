```python
import pandas as pd
import numpy as np
import random
# 设置全局Seed
np.random.seed(1)

# 文件读取：一种特殊类型的表格文件
train = pd.read_table('user_tag_query.10W.TRAIN',
                      encoding='gb18030',
                      header=None,
                      sep='\n',
                      names=['col'])
# 重新定义表格
train['ID'] = train['col'].apply(lambda x: x.split('\t')[0])
train['age'] = train['col'].apply(lambda x: x.split('\t')[1])
train['Gender'] = train['col'].apply(lambda x: x.split('\t')[2])
train['Education'] = train['col'].apply(lambda x: x.split('\t')[3])
# 内容：使用,进行拼接
train['Query'] = train['col'].apply(lambda x: ','.join(x.split('\t')[4:-1]))
# 对行别、年龄、教育程度进行过滤
train = train.drop(list(train[(train['age'] == '0') | (train['Gender'] == '0')
                              | (train['Education'] == '0')].index),
                   axis=0)
# 找到三种标签相同的人群进行gb 并提取信息
train_new = train.groupby(
    ['age', 'Gender', 'Education'],
    as_index=False).apply(lambda x: x['Query'].str.cat(sep=','))
```

```python
df = pd.DataFrame(train_new)
# 统计每种有多少个
c = train.groupby(['age','Gender','Education'],as_index = False).count()
```

```python
# 随机生成数据
def random_query(data, age, gender, edu):
    L = []
    df = pd.DataFrame()
    count1 = int(c[(c['age'] == str(age)) & (c['Gender'] == str(gender)) &
                   (c['Education'] == str(edu))]['ID'])
    count = count1 // 50  #if count1>100 else count1
    # 构建具体的词条表
    query = (data.loc[str(age), str(gender), str(edu)].values[0]).split(',')
    for i in range(count):
        # 从词条表种随机抽取（n条数据）
        L.append(random.sample(query, random.randint(18, len(query))))
    df['Query'] = L
    df['age'] = age
    df['Gender'] = gender
    df['Education'] = edu
    yield df
```

```python
T = pd.DataFrame()
for age in range(1, 7):
    for gender in range(1, 2):
        for edu in range(1, 7):
            T = T.append(random_query(df, age, gender, edu).__next__())
```

```python
T.to_csv('user_tag_T60.csv', encoding='gb18030')
```
